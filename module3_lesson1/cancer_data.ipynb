```python
#Importing Dependencies
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#Loading Dataset into Dataframe
df = pd.read_csv('cancer_data.csv')
#Check
df
```

```python
#DATA CLEANING

#check properly labelling of columns
#check the columns data type
#check for missing values,work on it by either to drop them or fill them up
#check for rows with duplicates and drop them
#check for the structure of the data set
```

```python
#check properly labelling of columns
df.columns #Columns are properly labeled i.e no missing value inbetween
```

```python
#check the columns data type
df.dtypes# Columns are integer,object and float datatypes but mostly Float datatype
```

```python
#check for missing values,work on it by either to drop them or fill them up
df.isna().sum()# there are missing values
```

```python
#Working on the missing values Using fillna to fill up empty spaces in the columns
#columns with the missinng values are float datatype filling them up with the average(mean)
# Fill up empty spaces  for 'texture_mean' column using the mean
a = df['texture_mean'].mean()
df['texture_mean'].fillna(a, inplace=True)

# Fill up empty spaces  for 'smoothness_mean' column using the mean
b = df['smoothness_mean'].mean()
df['smoothness_mean'].fillna(b, inplace=True)

# Fill up empty spaces  for 'symmetry_mean' column using the mean
c = df['symmetry_mean'].mean()
df['symmetry_mean'].fillna(c, inplace=True)

# Fill up empty spaces  for 'texture_SE' column using the mean
d = df['texture_SE'].mean()
df['texture_SE'].fillna(d, inplace=True)

# Fill up empty spaces for 'smoothness_SE' column using the mean
e = df['smoothness_SE'].mean()
df['smoothness_SE'].fillna(e, inplace=True)

# Fill up empty spaces  for 'symmetry_SE' column using the mean
f = df['symmetry_SE'].mean()
df['symmetry_SE'].fillna(f, inplace=True)

# Fill up empty spaces  for 'texture_max' column using the mean
g = df['texture_max'].mean()
df['texture_max'].fillna(g, inplace=True)

# Fill up empty spaces  for 'smoothness_max' column using the mean
h = df['smoothness_max'].mean()
df['smoothness_max'].fillna(h, inplace=True)

# Fill up empty spaces  for 'symmetry_max' column using the mean
i = df['symmetry_max'].mean()
df['symmetry_max'].fillna(i, inplace=True)

```

```python
#Check for missing values again
df.isna().sum()
```

```python
#check for rows with duplicates
df[df.duplicated()]
```

```python
#Drop Duplicates
df.drop_duplicates(inplace= True)
```

```python
#Save clean dataframe
cleaned_df = df
cleaned_df.to_csv("Final_df.csv", index=False)
```

```python
#Understanding the distribution of a column(radius_mean)
sns.kdeplot(cleaned_df["radius_mean"]); #radius_mean is more distributed between 10 and 15 and less around 20
```

```python
#Plot and customize a correlations heatmap for cleaned_df.
corr=cleaned_df.corr()
# use numpy function to mask the data to create a matrix of booleans
ones = np.ones_like(corr, dtype='bool') # by default returns ones for all the data

# use another numpy function to get just the upper triangle of the boolean array
mask = np.triu(ones)

# adjust mask and correlations 
mask = mask[1:, :-1]
corr = corr.iloc[1:,:-1].copy()

# generate customized heatmap of correlations
fig, ax = plt.subplots(figsize=(15,15))
sns.heatmap(corr, mask=mask, annot=True,fmt=".1f",cmap='YlOrRd', linewidths=.10, cbar_kws={'shrink':.99})
ax.xaxis.tick_bottom()
plt.title("Correlations heatmap of Cancer Data \n".upper(), fontdict={'fontsize': 25}, loc='center')
plt.show()
```

```python
#Insights on the the Relationship between Columns with commom values(_mean),(_SE),(_max) or different Specific Values


plt.figure(figsize=(15,10)) 
#Understanding the correlation of columns with the mean values
plt.subplot(2,2,1)
sns.regplot(x = cleaned_df["radius_mean"], y =cleaned_df["area_mean"] );
plt.title('Correlation between radius_mean \n and area_mean (Cancer data)'.upper())

#Understanding the Correlation of columns with the SE values 
plt.subplot(2,2,2)
sns.regplot(x = cleaned_df["smoothness_SE"], y =cleaned_df["symmetry_SE"] );
plt.title('Correlation between smoothness_SE \n and symmetry_SE (cancer Data)'.upper())

#Understanding the Correlation of columns with the different specific values
plt.subplot(2,2,3)
sns.regplot(x = cleaned_df["radius_max"], y =cleaned_df["fractal_dimension_SE"], color='green' );
plt.title('Correlation between radius_max \n and fractal_dimension_se(cancer Data)'.upper())

#Understanding the Correlation of columns with the different specific values
plt.subplot(2,2,4)
sns.regplot(x = cleaned_df["smoothness_SE"], y =cleaned_df["area_mean"] , color='green');
plt.title('Correlation between smoothness_se \n and area_mean (cancer Data)'.upper())

plt.subplots_adjust(wspace=0.5, hspace=0.5, bottom=0.1, top=0.9)
plt.show()
```

```python
#Columns with values in common(_mean),(_SE),(_max)are positively correlated with each other they differ between high or low
#Columns with different specific values are negatively correlated or no correlation with each other.
```
