```python
#Import Dependencies
import pandas as pd
```

```python
#Load csv dataset Using pd.read_csv()
df = pd.read_csv('all_alpha_08.csv')
df.head()
```

```python
#Check Dataframe datatypes
df.dtypes #Object and float datatypes mostly object
```

```python
#check for missing/null values,work on it by filling up or drop them
df.isna().sum()# there are missing/null values
```

```python
#Drop Null values
df.dropna(inplace=True)
```

```python
#Check again for missing/null values
df.isna().sum()#No missing/null values
```

```python
#Extract integer from string in the "Cyl"column using Pandas Series.str.extract()
df['Cyl'] = df['Cyl'].str.extract('(\d+)').astype(int) #regex(\d+)-Extract/capture one or more digit (0 - 9)
```

```python
#Confirm the Change
df['Cyl'].value_counts()
```
