```python
#Import Depndencies
import  twint
import pandas as pd
import numpy as np
import re
from textblob import TextBlob
import requests
import nest_asyncio
nest_asyncio.apply()
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
candidate_name={"Peter Obi": "\"Obidients\" OR \"ObiDatti\"",
    "Bola Ahmed Tinubu": "\"BAT\" OR \"Tinubu\" OR \"Jagaban\"",
    "Atiku Abubakar":"\"Atiku\" OR \"Atiku Abubakar\""}
```

```python
def twintConfig(search_string):
    c = twint.Config()
    c.Search = search_string[1]
    c.Limit=500
    c.Lang = 'en'
    c.Pandas = True
    c.Pandas_au = True          
    c.Pandas_clean=True
    c.Since='2022-01-01'
    c.Until='2022-12-31'
    twint.run.Search(c)
```

```python
def Run_Twint(search_vals):
    
    #Set empty dataframe for join
    out_df= pd.DataFrame()
    
    for candidate in search_vals.items():
        print ("running for search item: "+candidate[0]+"\n")
        print ("Search string: "+candidate[1]+"\n")
        
        #run twint
        twintConfig(candidate)
        
        #get dataframe
        df_all = twint.storage.panda.Tweets_df
        
        #join Dataframes and create 'Candidate' column
        df_all["Candidate"]= candidate[0]
        out_df = pd.concat([out_df,df_all])
        
    return out_df
```

```python
df_all = Run_Twint(candidate_name)
```

```python
def clean_text(text): 
    pat1 = r'@[^ ]+'                   #@signs
    pat2 = r'https?://[A-Za-z0-9./]+'  #links
    pat3 = r'\'s'                      #floating s's
    pat4 = r'\#\w+'                     # hashtags
    pat5 = r'&amp '
    pat6 = r'[^A-Za-z\s]'         #remove non-alphabet
    combined_pat = r'|'.join((pat1, pat2,pat3,pat4,pat5, pat6))
    text = re.sub(combined_pat,"",text).lower()
    return text.strip()
```

```python
df_all["tweet"] = df_all["tweet"].apply(clean_text)
```

```python
df_all = df_all [-(df_all["tweet"]=="")]
```

```python
df_all["tweet"]
```

```python
print("Running sentiment Analysis")

# creating two new columns(polarity and subjectivity)
def getSubjectivity(text):
    return TextBlob(text).sentiment.subjectivity
def getPolarity(text):
    return TextBlob(text).sentiment.polarity

df_all["Subjectivity"] = df_all['tweet'].apply(getSubjectivity)
df_all["Polarity"] = df_all['tweet'].apply(getPolarity)
```

```python
tweets_df=df_all.copy()
```

```python
tweets_df.tail(5)
```

```python
def getAnalysis(text):
    if text < 0:
        return 'negative'
    elif text == 0:
        return 'neutral'
    else:
        return 'positive'
           
```

```python
tweets_df['Analysis'] = tweets_df['Polarity'].apply(getAnalysis)
```

```python
tweets_df.tail(5)
```

```python
#Peter obi sentiment
peter_obi = tweets_df.loc[tweets_df['Candidate'] == 'Peter Obi']
peterobi_analysis= peter_obi['Analysis'].groupby(peter_obi['Analysis']).size().reset_index(name='counts')
peterobi_analysis
```

```python
#Bola Ahmed Tinubu sentiment
Bola_Ahmed_Tinubu = tweets_df.loc[tweets_df['Candidate'] == 'Bola Ahmed Tinubu']
BolaAhmedTinubu_analysis= Bola_Ahmed_Tinubu['Analysis'].groupby(Bola_Ahmed_Tinubu['Analysis']).size().reset_index(name='counts')
BolaAhmedTinubu_analysis
```

```python
#Atiku Abubakar sentiment
Atiku_Abubakar =  tweets_df.loc[tweets_df['Candidate'] == 'Atiku Abubakar']
AtikuAbubakar_analysis= Atiku_Abubakar['Analysis'].groupby(Atiku_Abubakar['Analysis']).size().reset_index(name='counts')
AtikuAbubakar_analysis
```

```python
#Grouping positive sentiments of candidates
positive_candidates = tweets_df.loc[tweets_df['Analysis'] == 'positive']
positive_tweets = positive_candidates['Candidate'].groupby(positive_candidates['Candidate']).size().reset_index(name='counts')
positive_tweets
```

```python
#plotting the results
plt.figure(figsize=(20,6))
sns.barplot(data = positive_tweets, x= 'Candidate', y ='counts', palette = 'pastel')

#setting the properties for the chart
plt.title('Presidential Candidates Polarity', fontsize = 20 ) 
plt.xlabel('Candidates')
plt.ylabel('Positive tweets')

#saving the chart
plt.show()
```

Peter Obi is the Presidential Candidate with the most positive polarity among the remaining candidates on Twitter.
