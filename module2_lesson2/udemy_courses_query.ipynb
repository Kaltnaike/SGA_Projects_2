```python
#Importing csv File into Python Environment Using Pandas
#Import Dependencies
import pandas as pd
import sqlalchemy
from sqlalchemy import create_engine

# Create Engine to connect to sqlite Database
engine = create_engine("sqlite+pysqlite:///:memory:")

#Load csv File into python environment using pd.read_csv()
df = pd.read_csv('udemy_courses.csv')

# load the Dataframe into Table in the Database
df.to_sql('udemy_data', engine)

# Connect to Engine
db_engine = engine.connect()

print("All Successful and Query Ready!")
```

```python
# Defining a Function to make it reusable to query different data from the Database
def query_db(query: str, db_conn: sqlalchemy.engine.base.Connection) -> pd.DataFrame:
    df = pd.read_sql(query, db_conn)
    return df

#Check
query_db("SELECT * FROM udemy_data LIMIT 5", db_engine)

```

```python
#Write a query for paid courses, showing the  course with the highest number of subscribers per month
query_db("""SELECT course_title, MAX(num_subscribers) AS highest_subscription,strftime('%m', published_timestamp) AS month
            FROM udemy_data 
            WHERE is_paid = 1 
            GROUP BY strftime('%m', published_timestamp)""", db_engine)
```

```python
#The Query Provided the Paid Course with the Highest number of subscribers per month(in each month from Janaury to December)
```

```python
#Write a query for paid courses, showing the course with the highest sales per month
query_db("""SELECT course_title,MAX(price*num_subscribers) AS sales,strftime('%m', published_timestamp) AS month
            FROM udemy_data
            WHERE is_paid = 1
            GROUP BY strftime('%m', published_timestamp)""", db_engine)
```
